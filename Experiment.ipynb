```python
import pandas as pd
import matplotlib.pyplot as plt

traject = pd.read_csv('051 daa0 2022 03 16 12 41 33 gps.dat',sep =",",skiprows=1)
traject[' latitude'] = traject[' latitude'] - 52.020
traject[' longitude'] = traject[' longitude'] - 4.268

Wt_   = list(traject['time'])
Wlat_ = list(traject[' latitude'])
Wlon_ = list(traject[' longitude'])

Wt   = []
Wlat = []
Wlon = []

time1 = -100
for index in range(0,len(Wt_)):
    time2 = round(Wt_[index])
    if time2 >= 100 and time2 < 368:
        if time1 != time2:
            Wt.append(round(Wt_[index]))
            time1 = time2
            Wlat.append(Wlat_[index])
            Wlon.append(Wlon_[index])

print(len(Wt))
print(Wt)

plt.figure()
fig,ax = plt.subplots()
plt.plot(Wt,Wlat,label= 'Watch lat')
plt.plot(Wt,Wlon,label= 'Watch lon')
plt.xlabel("time [sec]")
plt.ylabel("GPS decimal")
plt.legend()

plt.figure()
fig,ax = plt.subplots()
plt.plot(Wlat,Wlon,label= 'Watch lat vs lon')
plt.xlabel("latitude decimal")
plt.ylabel("longitude decimal")
plt.legend()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

def get_seconds(time_str):
    hh, mm, ss = time_str.split(':')
    sss,ns = ss.split(',')
    return int(hh) * 3600 + int(mm) * 60 + int(sss)

col_list = ["tagId","timestamp","dateTime","loc(x)","loc(y)"]
uwb1 = pd.read_csv("test1.csv",usecols=col_list)
uwb1 = uwb1[uwb1["tagId"] == 26694]

Ut_ = list(uwb1['dateTime'])
Ux_ = list(uwb1['loc(x)'])
Uy_ = list(uwb1['loc(y)'])

Ux = []
Uy = []
Ut = []

time1 = 0
for index in range(0,len(Ut_)):
    time2 = get_seconds(Ut_[index]) - 45693
    if time2 >= 100 and time2 < 400:
        if time1 != time2:
            Ux.append(Ux_[index])
            Uy.append(Uy_[index])
            Ut.append(time2)
            time1 = time2

print(len(Ux))
print(len(Uy))
print(len(Ut))
print(Ut)

plt.figure()
fig,ax = plt.subplots()
plt.plot(Ut,Ux,label= 'UWB x')
plt.plot(Ut,Uy,label= 'UWB y')
plt.xlabel("time [sec]")
plt.ylabel("x or y [mm]")
plt.legend()
```

```python
#
# Take the inner join of U and W, where len(U) < len(W) and Ut = Wt
#
# resulting in Mt, Mx, My, Mlat, Mlon
#
Mt    = []
Mx    = []
My    = []
Mlat  = []
Mlon  = []

for index in range(0,len(Ut)):
    for index2 in range(0,len(Wt)):
        if Ut[index] == Wt[index2]:
            #print(Ut[index], Wt[index2], index, index2, Ux[index], Wlon[index])
            Mt.append(Ut[index])
            Mx.append(Ux[index]/10000000)
            My.append(Uy[index]/10000000)
            Mlat.append(Wlat[index2])
            Mlon.append(Wlon[index2])

print(len(Ut))
#print(Mx)

plt.figure()
fig,ax = plt.subplots()
plt.plot(Mt,Mx,label= 'UWB x')
plt.plot(Mt,My,label= 'UWB y')
plt.plot(Mt,Mlat,label= 'Watch lat')
plt.plot(Mt,Mlon,label= 'Watch lon')
plt.xlabel("time [sec]")
plt.ylabel("x or y [meters * 10,000] or GPS decimal")
plt.legend()

```

```python
A = []
B = []
A.append(pd.read_csv('010 dae4 2022 03 16 12 12 56 gps.dat', sep =",", skiprows=1))
A.append(pd.read_csv('020 db3a 2022 03 16 12 15 34 gps.dat', sep =",", skiprows=1))
A.append(pd.read_csv('030 dae4 2022 03 16 12 13 23 gps.dat', sep =",", skiprows=1))
A.append(pd.read_csv('040 dac4 2022 03 16 12 13 36 gps.dat', sep =",", skiprows=1))

T = [-1717, -1559, -1690, -1677]
Q = [[52.020161,4.268634],[52.020093,4.268767],[52.019981,4.268616],[52.020048,4.268484]]

index = 0
for beacon in A:
    beacon['time'] = beacon['time'] + T[index]
    beacon['latitude'] = beacon['latitude'] - 52.020
    beacon['longitude'] = beacon['longitude'] - 4.268
    B.append(beacon)
    Q[index][0] -= 52.020
    Q[index][1] -=  4.268
    index += 1

#print(B[0]['time'])

Bt      = [[],[],[],[]]
Blat    = [[],[],[],[]]
Blon    = [[],[],[],[]]
Bdiflat = [[],[],[],[]]
Bdiflon = [[],[],[],[]]
Bdst    = [[],[],[],[]]

#
# Take out timeslot 100 to 368 seconds from beacon coordinates
#
from math import sqrt

indexb = 0
for beacon in B:
    time1 = -100
    for index in range(0,len(beacon)):
        time2 = round(beacon['time'][index])
        if time2 >= 100 and time2 < 368:
            if time1 != time2:
                Bt[indexb].append(round(beacon['time'][index]))
                time1 = time2
                Blat[indexb].append(beacon['latitude'][index])
                Blon[indexb].append(beacon['longitude'][index])
                dlat = Q[indexb][1] - beacon['latitude'][index]
                dlon = Q[indexb][0] - beacon['longitude'][index]
                Bdiflat[indexb].append(dlat)
                Bdiflon[indexb].append(dlon)
                Bdst[indexb].append(sqrt(dlat*dlat + dlon*dlon))
    indexb += 1

#for index in range(0,4):
#    print(len(Bt[index]))
#print(Bt)

plt.figure()
fig,ax = plt.subplots()
for index in range(0,4):
    slabel = "Blat" + str(index+1)
    plt.plot(Bt[index], Blat[index], label=slabel)
for index in range(0,4):
    slabel = "Blon" + str(index+1)
    plt.plot(Bt[index],Blon[index],label=slabel)
plt.xlabel("time [sec]")
plt.ylabel("GPS [decimal]")
plt.legend()

plt.figure()
fig,ax = plt.subplots()
for index in range(0,4):
    slabel = "Bdst" + str(index+1)
    plt.plot(Bt[index], Bdst[index], label=slabel)
plt.xlabel("time [sec]")
plt.ylabel("sqrt(dlat+dlon) [gps decimal]")
plt.legend()

plt.figure()
fig,ax = plt.subplots()
for index in range(0,4):
    slabel = "Blat vs lon" + str(index+1)
    plt.plot(Blat[index], Blon[index], label=slabel)
plt.xlabel("latitude [gps decimal]")
plt.ylabel("longitude [gps decimal]")
plt.legend()

plt.figure()
fig,ax = plt.subplots()
for index in range(0,4):
    slabel = "Bdif lat vs lon" + str(index+1)
    plt.plot(Bdiflat[index], Bdiflon[index], label=slabel)
plt.xlabel("diff latitude [gps decimal]")
plt.ylabel("diff longitude [gps decimal]")
plt.legend()

```
